```python
print('Number of Male and Female in the dataset: \n', df.groupby('sex').size())
```

```python
plt.figure(figsize=(12,8))
df.groupby('sex').size().plot(kind='pie', label='', autopct = '%.1f')
plt.title('Male-Female Distrbution')
```

```python
plt.figure(figsize=(12,8))
df.groupby('heartdisease').size().plot(kind='pie', autopct = '%0.1f', label='')
plt.title('Heart Disease Distribution Pie Chart', size=20)
```

```python
df.groupby('chestpaintype').size().plot(kind='pie', label = 'Types of Chest Pain', autopct = '%.1f')
```

```python
plt.figure(figsize=(12,8))
df.groupby('restingecg').size().plot(kind='pie', autopct = '%0.1f', label='')
plt.title('Resting ECG', size=20)
```
